```python
select_category = driver.find_element_by_xpath('//div[@class="_p13n-zg-nav-tree-all_style_zg-browse-group__88fbz"]//div[34]//a').click()
```

```python
products = []
prices = []
titles = []
stars = []
```

```python
while True:
    driver.execute_script("""
    let element = document.getElementById('zg');
    
    element.scrollIntoView({block: "end", behavior: "smooth"});
    
    """)
    p = driver.find_elements_by_xpath('//div[@class="zg-grid-general-faceout"]/div/a[1]')
    p = [i.get_attribute('href') for i in p]
    price = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]/span[@class="a-size-base a-color-price"]')
    price = [i.text for i in price]
    t = driver.find_elements_by_xpath('//div[@class="_p13n-zg-list-grid-desktop_truncationStyles_p13n-sc-css-line-clamp-3__g3dy1"]')
    t = [i.text for i in t]
    for i in p:
        if i not in products:
            products.append(i)
    for i in price:
        if i not in prices:
            prices.append(i)
    for i in t:
        if i not in titles:
            titles.append(i)

    print(len(products))
    if len(products)==50:
        break
```
